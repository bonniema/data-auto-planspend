```python
#Combine Digital and Offline to one file
planspend_new_before = pd.concat([digital_new, other_new], ignore_index = True)
planspend_new_before.head()
```

```python

q = """

SELECT Custom_Quarter,
Custom_Month, 
Business_vs_Resi, 
Division,
Region,
Market, 
DMA_Code, 
Media, 
Media_Type, 
Input_Product_Code, 
Plan_Name,
sum(Allocated_Planned_Net_Cost) as Sum_of_Allocated_Planned_Net_Cost,
Custom_Week
FROM planspend_new_before
GROUP BY 
Custom_Quarter,
Custom_Month, 
Business_vs_Resi, 
Division,
Region,
Market, 
DMA_Code, 
Media, 
Media_Type, 
Input_Product_Code, 
Plan_Name,
Custom_Week
ORDER BY
Custom_Week;


"""


```

```python
results = pysqldf(q)
```

```python
results.head()
```

```python
results['Year'] = '2019'
planspend_new = results
```

```python
planspend_new.head()
```

```python
def new_quarter(x):
    
    if x =='01JAN19':
        a = 'Q1'
    if x =='01APR19':
        a = 'Q2'
    if x =='01JUL19':
        a = 'Q3'
    if x =='01OCT19':
        a = 'Q4'
    return a
```

```python
planspend_new['Custom_Quarter']=planspend_new['Custom_Quarter'].apply(new_quarter)
```
